```python
# Open and read the JSON file
with open('/Users/rahnumatarannum/bk_defenders_surveillance_capstone/Outputs/KDE_sampling.json', 'r') as f:
    data = json.load(f)

print(type(data),len(data))
```

```python
features = []
for item in data[2]:
    geometry = shape(item['geojson'])
    properties = item['results']
    features.append({"geometry": geometry, "properties": properties})

nta_surv = gpd.GeoDataFrame.from_features(features)
nta_surv.head()
```

```python
filepath = "/Users/rahnumatarannum/bk_defenders_surveillance_capstone/Outputs"
nta_surv.to_file(f"{filepath}/kde_final_output.shp", driver='ESRI Shapefile')
```
